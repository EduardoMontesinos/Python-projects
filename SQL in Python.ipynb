```python
sql("""
    
    SELECT 
        handling.nombre,
        SUM(repostajes.litros) as 'Litros totales'
    FROM handling
      
    LEFT JOIN vehiculos ON
    handling.id=vehiculos.id_handling
    
    LEFT JOIN repostajes ON
    vehiculos.id=repostajes.id_vehiculo
    
    -- WHERE nombre="GROUND FORCE"
    
    GROUP BY handling.nombre

""")
```

```python
sql("""

    SELECT
        vehiculos_filtro.*,
        repostajes.id_gasolinera,
        repostajes.litros,
        repostajes.precio_litro,
        repostajes.fecha
    FROM (
        SELECT
        *
        FROM(
        SELECT 
            id as handling_id,
            nombre
        FROM handling

        WHERE nombre="GROUND FORCE"
        ) handling_filtro

        LEFT JOIN vehiculos ON
        handling_filtro.handling_id=vehiculos.id_handling

        WHERE vehiculos.tipo='Escalera'
    
    ) vehiculos_filtro
    
    LEFT JOIN repostajes ON
    vehiculos_filtro.id=repostajes.id_vehiculo


""")
```
